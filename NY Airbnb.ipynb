### 1) Neighbourhood group vs no. of listings

Looks like Manhattan and Brooklyn tops in the number of Airbnb listings

```python
df.groupby('room_type')['id'].count()
```

```python
df.groupby(['neighbourhood_group','room_type'])['id'].count().reset_index()
```

```python
value = df.groupby(['neighbourhood_group','room_type'])['id'].count().reset_index()
#value.rename(columns={'id':'Number of listings'},inplace=True)
data = value.pivot(index='neighbourhood_group', columns='room_type', values='id')
data
```

```python
data.plot(kind='bar',figsize=(10,6))
plt.title('Borough vs Number of Listings');
plt.ylabel('Number of Listings');
```

```python
data.plot(subplots=True, figsize=(8,6))
```

As you can see in the above plot, Manhattan and Brooklyn has high number of Entire Home/Apt and Private room

### 2) Room Type vs number of listings

### 5) Room Type vs Availability (days)

```python
df[['room_type','availability_365']]
```

### 6 ) host_id vs number of listings

```python
hosts = df.groupby(['host_id'])['id'].count().sort_values(ascending=False).head(10)
hosts.plot.bar(color='g');
plt.title('Top 10 Hosts per number of listings');
plt.ylabel('Number of listing');
```

```python
hosts = df.groupby(['host_id'])['id'].count().sort_values(ascending=False).reset_index().head(10)
hosts.rename(columns={'id':'Number of listings'},inplace=True)

plt.figure(figsize=(8, 6));
sns.barplot(hosts['host_id'],hosts['Number of listings'],palette='rocket_r');
plt.title('Top 10 Hosts per number of listings');
plt.xticks(rotation=70);
```
